Cetegories average_vote column
We would cut the vote_Average values and make 4 categories. popular average below_abg not_popular to describe it more using categories_col() function provide above


```python
def categorize_col(df,col,labels):
    edges = [df[col].describe()['min'],
             df[col].describe()['25%'],
             df[col].describe()['50%'],
             df[col].describe()['75%'],
             df[col].describe()['max']]
    df[col] = pd.cut(df[col], edges, labels = labels, duplicates = 'drop')
    return df
```

```python
labels = ['not_popular', 'below_avg', 'average','popular']

categorize_col(df,'vote_average', labels)

df['vote_average'].unique()
```

### Which genres have the highest average ratings?

```python
genre_avg_rating = df.groupby('genres')['rating'].mean().sort_values(ascending=False).reset_index()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=genre_avg_rating, x='rating', y='genres', color = "green")
plt.title('Average Ratings by Genres')
plt.xlabel('Average Ratings')
plt.ylabel('Genres')
plt.tight_layout()
plt.show()
```
